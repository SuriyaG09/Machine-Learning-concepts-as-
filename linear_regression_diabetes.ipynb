#Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Linear regression predicts relationships between feature X and target variable y

```python
X, Y = datasets.load_diabetes(return_X_y=True)

print(X.shape)
print(X[0])
```

#Target:
Column 11 is a quantitative measure of disease progression one year after baseline

```python
X = X[:,2] # taking only BMI from column 3 as the parameter for the model
X = X.reshape((-1,1))
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, Y, test_size=0.33)
```

# Loading the linear model

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

# Prediction

```python
y_pred = model.predict(X_test)
```

# Scatter plot for prediction

```python
plt.scatter(X_test, y_test, color='darkviolet')
plt.plot(X_test, y_pred, color = 'black', linewidth = 3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

## Using s6 glu, blood sugar level as the data for linear regression

```python
X, Y = datasets.load_diabetes(return_X_y=True)
```

```python
X = X[:,9]
X = X.reshape((-1,1))
```

```python
# train and test

xtr, xte, ytr, yte = model_selection.train_test_split(X,Y,test_size = 0.33)
```

```python
# model creation

model1 = linear_model.LinearRegression()

model1.fit(xtr,ytr)
```

```python
# predict

ypre = model1.predict(xte)
```

```python
#plot

plt.scatter(xte, yte)
plt.plot(xte, ypre, color = 'red', linewidth='3')
plt.xlabel('Scaled Blood Sugar level')
plt.ylabel('Disease Progression')
plt.title('Diabetes Progression against blood sugar level')
plt.show()
```
